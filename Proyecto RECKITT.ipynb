# 4. Transformaciones necesarias

## Sales_Week
### (DataFrame unido)

```python
#Limpieza del DataFrame eliminando duplicados y transformacion de tipo de datos
sales_week = sales_week.dropna()
sales_week = sales_week.drop_duplicates()
sales_week['MONTH'] = sales_week['MONTH'].astype(int)
sales_week['WEEK_NUMBER'] = sales_week ['WEEK_NUMBER'].astype(int)
sales_week['YEAR'] = sales_week['YEAR'].astype(int)

sales_week
```

```python
#Estandarizacion de datos de la columna ITEM_CODE
sales_week['ITEM_CODE'] = sales_week['ITEM_CODE'].astype(str)
sales_week['ITEM_CODE'] = sales_week['ITEM_CODE'].apply(lambda x: format(float(x), '.0f') if 'E+' in str(x) else x)
sales_week
```

## Prod_Sale
### (DataFrame unido)

```python
#Limpieza del DataFrame eliminando duplicados y transformacion de tipos de datos
prod_sale = prod_sale.dropna()
prod_sale = prod_sale.drop_duplicates()
prod_sale ['CATEGORY'] = prod_sale['CATEGORY'].astype(int)
prod_sale['ITEM_CODE'] = prod_sale['ITEM_CODE'].astype(str)

prod_sale
```

```python
#Estandarizacion de la columna WEEK
meses = {
    'ene': '01', 'feb': '02', 'mar': '03', 'abr': '04',
    'may': '05', 'jun': '06', 'jul': '07', 'ago': '08',
    'sep': '09', 'oct': '10', 'nov': '11', 'dic': '12'}
def convertir_mes(valor):
    if isinstance(valor, str) and '-' in valor:
        parte1, parte2 = valor.split('-')
        parte1 = parte1.lower().strip()
        
        if parte1 in meses:
            return f"{meses[parte1]}-{parte2}"
    
    return valor

prod_sale['WEEK'] = prod_sale['WEEK'].apply(convertir_mes)
prod_sale
```

```python
#Estandarizacion de la columna ITEM y ITEM_CODE
prod_sale['ITEM_CODE'] = prod_sale['ITEM_CODE'].astype(str)
prod_sale['ITEM_CODE'] = prod_sale['ITEM_CODE'].apply(lambda x: format(float(x), '.0f') if 'E+' in str(x) else x)
prod_sale['ITEM'] = prod_sale['ITEM'].astype(str)
prod_sale['ITEM'] = prod_sale['ITEM'].apply(lambda x: format(float(x), '.0f') if 'E+' in str(x) else x)
prod_sale
```

```python
# Union y limpieza de DataFrames unidos 
prod_sale_week = pd.concat([prod_sale,sales_week], axis = 1)
prod_sale_week = prod_sale_week.fillna(0)
prod_sale_week = prod_sale_week.loc[:, ~prod_sale_week.columns.duplicated() ]
prod_sale_week ['YEAR'] = prod_sale_week ['YEAR'].astype(int)
prod_sale_week ['MONTH'] = prod_sale_week ['MONTH'].astype(int)
prod_sale_week ['WEEK_NUMBER'] = prod_sale_week ['WEEK_NUMBER'].astype(int)
prod_sale_week
```

```python
final_prod = prod_sale_week.groupby(
    ['ITEM', 'BRAND','WEEK','ATTR1', 'REGION']).agg({'TOTAL_UNIT_SALES': ['sum', 'median', 'std'],'TOTAL_VALUE_SALES': 'sum',
                                                                   'TOTAL_UNIT_AVG_WEEKLY_SALES': 'mean'}).reset_index()

final_prod = final_prod.fillna(0)
final_prod
```

# 5. Guardar el archivo

```python
#Guardar el DataFrame Final
final_prod.to_csv("FINAL_PROD.csv")
```
