# Correccion del archivo csv de DIM_CALENDAR

```python
#Abrir el archivo que se busca corregir
import pandas as pd
df = pd.read_csv('DIM_CALENDAR.csv', encoding = 'ISO-8859-1' )
df
```

```python
#Correccion de la columna WEEK
def corregir_week(valor):
    try:
        meses = {
            'ene': 1, 'feb': 2, 'mar': 3, 'abr': 4,
            'may': 5, 'jun': 6, 'jul': 7, 'ago': 8,
            'sep': 9, 'oct': 10, 'nov': 11, 'dic': 12}

        if isinstance(valor, str) and '-' in valor:
            parte, year = valor.split('-')
            if parte.lower() in meses:
                mes = meses[parte.lower()]
            else:
                mes = int(parte)

            if 1 <= mes <= 12:
                return f'{mes:02d}-{year}'

        return valor

    except:
        return valor
```

```python
df['WEEK'] = df.apply(
    lambda x: f"{int(x['WEEK_NUMBER']):02d}-{str(x['YEAR'])[-2:]}"
    if pd.notnull(x['WEEK_NUMBER']) and pd.notnull(x['YEAR'])
    else x['WEEK'],
    axis=1)
```

```python
#Apliacion de la funcion sobre la columna a corregir
df['WEEK'] = df['WEEK'].apply(corregir_week)
df
```

```python
#Guardar el nuevo archivo corregido
df.to_csv('DIM_WEEK(1).csv', index = False)
```
