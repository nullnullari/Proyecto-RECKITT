### Limpieza de datos

```python
#Correccion de WEEK
def correccion_week(valor):
    try:
        meses = {
            'ene': 1, 'feb': 2, 'mar': 3, 'abr': 4,
            'may': 5, 'jun': 6, 'jul': 7, 'ago': 8,
            'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12
        }
        if isinstance(valor, str) and '-' in valor:
            parte, year = valor.split('-')
            if parte.lower() in meses:
                mes = meses[parte.lower()]
            else:
                mes = int(parte)
            if 1 <= mes <= 12:
                return f'{mes:02d}-{year}'
        return valor
    except:
        return valor
```

```python
#Correccion de week
week['WEEK'] = week.apply(
    lambda x: f'{int(x['WEEK_NUMBER']):02d}-{str(x['YEAR'])[-2]}'
    if pd.notnull(x['WEEK_NUMBER']) and pd.notnull(x['YEAR'])
    else x['WEEK'], axis = 1)
```

```python
#Aplicacion de la funcion sobre la columna con errores
week['WEEK'] = week['WEEK'].apply(correccion_week)
week
```
